```python
from sklearn.svm import SVR
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
data = load_boston()
data.data.shape,data.target.shape
```

```python
X_train,X_test,y_train,y_test = train_test_split(data.data,data.target,test_size=0.2)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
model = SVR()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
metrics.mean_absolute_error(y_test,y_pred)
```

```python
import pandas as pd
import numpy as np

def read_dataset(fname):
#     
    data = pd.read_csv(fname, index_col=0)
#     
    data.drop(['Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
#     
    lables = data['Sex'].unique().tolist()#
    data['Sex'] = [*map(lambda x: lables.index(x) , data['Sex'])]
#     ## sklearnlabelencoder
    lables = data['Embarked'].unique().tolist()
    data['Embarked'] = data['Embarked'].apply(lambda n: lables.index(n))
#     0
    data = data.fillna(0)
    return data
train = read_dataset("data/train.csv")
import warnings
warnings.filterwarnings("ignore")#
```

```python
from sklearn.model_selection import train_test_split#

y = train['Survived'].values#
X = train.drop(['Survived'], axis=1).values#

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)#
print("X_train_shape:", X_train.shape, " y_train_shape:", y_train.shape)
print("X_test_shape:", X_test.shape,"  y_test_shape:", y_test.shape)
```

```python
from sklearn.svm import SVC#
model = SVC()#
model.fit(X_train,y_train)#
print("train score:", model.score(X_train, y_train))#
print("test score:", model.score(X_test, y_test))#
```

## 

```python
from sklearn.model_selection import GridSearchCV#
from sklearn import metrics#
from sklearn.externals import joblib##
```

```python
degree=[1,2,3,4,5,6,7]
param_grid = {"degree":degree}##
```

```python
model = GridSearchCV(SVC(),param_grid,cv=5)#
```

```python
model.fit(X_train,y_train)#
```

```python
model.best_params_,model.best_score_
```

```python
model.score(X_test,y_test)
```

```python
metrics.accuracy_score(y_test,model.predict(X_test))
```

```python
metrics.f1_score(y_test,model.predict(X_test))#
```
